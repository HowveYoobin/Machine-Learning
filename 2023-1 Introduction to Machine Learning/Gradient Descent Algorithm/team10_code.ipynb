## **1. Create a plot showing the objective function within −5 ≤ $x$ ≤ 5**

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Define the objective function
def f(x):
  return np.sin(x) - np.cos(x)

# Create a range of x values
x = np.linspace(-5, 5, 1000)

# Calculate the function
y = f(x)

# Plot the objective function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Objective function')
plt.show()
```

## **2. Write a pseudo-code for your own algorithm**


1. Set Initial point $X_i$ between $ -5 \le x \le 5 $
2. Set the step size(learning rate) ${\alpha}$
3. set the steepest direction $S_k$
4. Set the max number of iterations $N$
5. Set the convergence threshold $\epsilon$


---
#### **Pseudocode**

**Input**: objective function (cost function) $f(x)$,  
Differentiated objective function $df(x)$,  
initial point $X_i$,  
Step Size ${\alpha}$,  
Steepest direction(partial derivative) $S_k$,  
Max number of Iterations $N$,  
Convergence threshold $\epsilon$


**Output**: $x$ values, $y$ values of $f(x)$ at each point after every loop

**set** $x$ = $X_i$  
**set** x_values = [ ]  
**set** y_values = [ ]  

**for** $i=1$ to $N$ **do**  
&nbsp;&nbsp;&nbsp;&nbsp; $x_{new}$ $\leftarrow$ $x - {\alpha}S_k$  
&nbsp;&nbsp;&nbsp;&nbsp; **if** $x_{new}$ $\lt$ -5 **then**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $x_{new}$ = -5   
&nbsp;&nbsp;&nbsp;&nbsp; **if** $x_{new}$ $\gt$ 5 **then**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $x_{new}$ = 5  
&nbsp;&nbsp;&nbsp;&nbsp; **end if**  
&nbsp;&nbsp;&nbsp;&nbsp; **if** $|f(x_{new} - f(x)|$ $\lt$ $\epsilon$ **then**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **exit for**  
&nbsp;&nbsp;&nbsp;&nbsp; **if** $f(x_{new})$ $\gt$ $f(x)$  **then**  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **exit for**  
&nbsp;&nbsp;&nbsp;&nbsp; **end if**  
&nbsp;&nbsp;&nbsp;&nbsp; x_values.append($x$)  
&nbsp;&nbsp;&nbsp;&nbsp; y_values.append($f(x)$)  
&nbsp;&nbsp;&nbsp;&nbsp; $x$ = $x_{new}$  
**end for**  
**return** x_values, y_values

## **3. Explain how you design a stopping criterion**

```python
# Define f(x) and f'(x)
def f(x):
  return np.sin(x) - np.cos(x)

def df(x):
  return np.cos(x) + np.sin(x)

# Define gradient descent function
def gradient_descent(f, df, xi, alpha, N, epsilon = 1e-8):
  x = xi
  x_values = []
  y_values = []
  for i in range(N):
    x_new = x - alpha * df(x)
    # x should be in the range of -5 and 5 (Stopping criterion)
    if x_new < -5:
      x_new = -5
    elif x_new > 5:
      x_new = 5
    # Stopping Criterion 1: Conversion threshold
    ''' If the Gradient Descent point gets close to the minimum, the step size will be very small, 
    but we want to stop if it makes very small movements that is not significant. 
    Therefore, we put convergence threshold to determine if we continue or not 
    when the gradient has changed less than certain number.'''
    if abs(f(x_new) - f(x)) < epsilon:
      break
    # Stopping Criterion 2: reject to accept worse situation
    ''' If the point of gradient descent gets close to the minimum, its y value will decrease. 
    To stop the point at the minimum of the graph, we should stop the point movement 
    before the f(x) value of updated x gets larger than that of x. 
    Therefore, we put comparation of f(x) value of updated x and x to determine whether it continues or not.'''
    if f(x_new) > f(x):
      break 
    x_values.append(x)
    y_values.append(f(x))
    x = x_new

  return x_values, y_values
```

#### **1. Suppose that you implement a stopping criterion where the algorithm terminates when the number of iterations exceeds the maximum number of iterations. When the following conditions are given, what is the optimal solution? Do you think that it is a global optimum within the pre-determined range? If not, what is your strategy to improve the fidelity of the solution provided by the algorithm?**

  * Initial point $x$ = 3  
  * Step size = 0.05  
  * The maximum number of iterations = 500

```python
''' Gradient_descent function with stopping critrion where the algorithm terminates 
when the number of iterations exceeds the maximum number of iterations'''
def gradient_descent_1(f, df, xi, alpha, N):
  x = xi
  x_values = []
  y_values = []
  for i in range(N):
    x_new = x - alpha * df(x)
    # x should be in the range of -5 and 5 (Stopping criterion)
    if x_new < -5:
      x_new = -5
    elif x_new > 5:
      x_new = 5
    # Accumulate the points for later plotting
    x_values.append(x)
    y_values.append(f(x))
    # Update x
    x = x_new
  mini = y_values.index(min(y_values))
  return x_values, y_values, y_values[mini]
```

```python
## Discussion 1
# Initial point
xi = 3
# The maximum number of iterations
N = 500
# Step size
alpha = 0.05

# Measure the start time of the execution
start_time = time.time()

# Implement the gradient_descent function
x_coordinates,y_coordinates, mini_y = gradient_descent_1(f, df, xi, alpha, N)

# Measure the end time of the execution
end_time = time.time()
# Measure the total execution time
exe_time = end_time - start_time
```

```python
# Plot a graph
x = np.linspace(-5, 5, 1000)
plt.plot(x, f(x))
for i in range(len(y_coordinates)):
  if y_coordinates[i] == min(y_coordinates):
    plt.scatter(x_coordinates[i], y_coordinates[i], c='red')
  else:
    plt.scatter(x_coordinates[i], y_coordinates[i], c='blue')
plt.hlines(mini_y, -5, 5, colors = "red", linestyle ='--')
plt.title("Gradient Descent")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# Print the minimum point
print(f"Minimum: x= {x_coordinates[y_coordinates.index(mini_y)]}, y= {mini_y}")
# Print the execution time
print(f"Execution time: {exe_time}")
```

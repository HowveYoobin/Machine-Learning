```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

## Problem
Find the minimum point of the objective function within -5<=x<=5 
- f(x)= sinx-cosx

```python
start=-5
end=5

x = np.linspace(start, end, 1000)
y = np.sin(x) - np.cos(x)

plt.plot(x, y)

plt.title("sin(x) - cos(x)")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

## Question1

```python
# gradient descent 함수 정의
def gradient_descent(f, df, x_init, stepsize, max_iter):
    # 초기값 설정
    x = x_init
    x_list = [x]
    y_list = [f(x)]
    
    # gradient descent 알고리즘 반복
    for i in range(max_iter):
        x = x - stepsize * df(x)
        
        # x가 범위 내에 있는지 확인
        if x < -5:
            x = -5
        elif x > 5:
            x = 5
        
        x_list.append(x)
        y_list.append(f(x))
        
        # 수렴 여부 확인
        if abs(y_list[-1] - y_list[-2]) < 1e-8:
            break
    
    # 최솟값과 그 위치 출력
    min_idx = np.argmin(y_list)
    print("최솟값:", y_list[min_idx])
    print("위치:", x_list[min_idx])
    
    return x_list, y_list

# 함수 및 미분 함수 정의
def f(x):
    return np.sin(x) - np.cos(x)

def df(x):
    return np.cos(x) + np.sin(x)

# gradient descent 알고리즘 실행
x_init = 3
stepsize = 0.05
max_iter = 500

x_list, y_list = gradient_descent(f, df, x_init, stepsize, max_iter)

# 그래프 그리기
x = np.linspace(-5, 5, 1000)
y = f(x)

plt.plot(x, y)
plt.scatter(x_list, y_list, c='r')
plt.title("Gradient Descent")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# 초기값을 여러개로 설정하여 전역최솟값을 확인한다.
```
